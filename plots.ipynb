```python
%matplotlib inline
from landscapegenerator import landscapeGenerator
from nkagent import nkAgent
from graphgenerator import graphGenerator
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
import time
import pickle
```

```python
def plot_all(result, title):
    fig, ax = plt.subplots(1,3, sharex='row', figsize = (18,5))
    s_model = []
    for i in range(len(result.keys())):
        s_ass = []
        for a in range(3):
            s_it = []
            for it in range(100):
                s_it.append(result[f'model_{i}']['assignments'][f'assignment_{a}']['iterations'][f'iter_{it}']['score'])
            s_ass.append(np.mean(np.array(s_it), axis = 0))
        s_model.append(np.array(s_ass))    
    ax[0].plot(np.mean(np.array(s_model)[:10], axis = 0).T, linewidth = 1)
    ax[1].plot(np.mean(np.array(s_model)[10:20], axis = 0).T, linewidth = 1)
    ax[2].plot(np.mean(np.array(s_model)[20:], axis = 0).T, linewidth = 1)
    ax[0].set_title('Complexity low, k = 2')
    ax[1].set_title('Complexity med, k = 4')
    ax[2].set_title('Complexity high, k = 8')    
    fig.suptitle(title)
    plt.legend(['exp_core', 'exp_peri', 'random'])
    
    return fig
```

# Base Model - 100 iterations

```python
i = 0
probs = [0.5,0.6,0.7,0.8,0.9]
for p in probs:
    title = f'Base Model with risk = 0, dist = 1 and p_exp = {p}'
    with open(f'data/base_model_{i}', 'rb') as f:
        data = pickle.load(f)
        
    plot_all(data, title)
    i+=1
```
